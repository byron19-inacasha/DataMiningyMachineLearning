## (u) Inspect which were the hottest days, but also with the highest ozone values.

```r
# Ordenar el conjunto de datos por temperatura en orden descendente
sorted_data <- airquality[order(airquality$Temp, decreasing = TRUE), ]

# Obtener el día más caliente
hottest_day <- sorted_data[1, ]

# Filtrar los días con el mismo valor de temperatura más alto
hottest_days <- sorted_data[sorted_data$Temp == hottest_day$Temp, ]

# Ordenar los días filtrados por el valor de ozono en orden descendente
sorted_hottest_days <- hottest_days[order(hottest_days$Ozone, decreasing = TRUE), ]

# Imprimir los resultados
print(sorted_hottest_days)
```

## (w) For each month, obtain the minimum and the maximum temperature registered in Celsius

```r
# Convertir la temperatura a Celsius
airquality$TempC <- (airquality$Temp - 32) * (5/9)

# Calcular la temperatura mínima y máxima para cada mes
temperature_stats <- aggregate(TempC ~ Month, data = airquality, FUN = function(x) c(min = min(x), max = max(x)))

# Imprimir los resultados
print(temperature_stats)
```

## (x) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.


```r
# Convertir la temperatura a Celsius
airquality$TempC <- (airquality$Temp - 32) * (5/9)

# Calcular el promedio de los parámetros por mes
averages <- aggregate(cbind(TempC, Wind, Solar.R, Ozone) ~ Month, data = airquality, FUN = mean)

# Imprimir los resultados
print(averages)
```

## (y) What values did you obtain regarding ozone and solar radiation attributes? Why? Make the necessary change so that you get the average of the registered values.

```r
# Filtrar los valores faltantes en ozono y radiación solar
filtered_data <- airquality[complete.cases(airquality$Ozone, airquality$Solar.R), ]

# Calcular el promedio de ozono y radiación solar
ozone_avg <- mean(filtered_data$Ozone)
solar_avg <- mean(filtered_data$Solar.R)

# Imprimir los resultados
print(ozone_avg)
print(solar_avg)
```
